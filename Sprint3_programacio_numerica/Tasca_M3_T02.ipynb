* #### Explica breument quines variables hi ha
Unnamed:

**Información del avión**
* UniqueCarrier: siglas compañia
* FLightNum: número de vuelo
* TailNum: número de avión

**Información, destino, origen y distancia del vuelo** 
* Origin: origen del vuelo
* Dest: destino del vuelo
* Distance: distacia del vuelo

**Fecha:** 
* Year:año del vuelo
* Month: mes del vuelo
* DayofMont:día del vuelo
* DayOfWeek: día de la semana del vuelo 

**Información horaria,tiempos y retrasos**
* DepTime: hora de salida real. (hhmm)
* CRSDepTime: hora de salida programada(hhmm)
* ArrTime: hora de llegada real(hhmm)
* CRSArrTime: hora de llegada progama (hhmm) 
* ActualElapsedTime:  tiempo total real entre la salida y la llegada del vuelo (minutos)
* CRSElapsedTime: tiempo total programado entre la salida y llegada del vuelo (minutos)
* AirTIme: tiempo de vuelo (minutos)
* ArrDelay: retraso en la llegada. (minutos)
* DepDelay: retraso en la salida. (minutos)
* TaxiIn: período desde el momento en el que el avión aterriza hasta que aparca en la puerta (minutos)
* TaxiOut: período desde el momento en el que el avión abandona la puerta hasta que despega(minutos)

**Tipos de retraso**
* Diverted: vuelo desviado, 1=sí 0=no
* CarrierDelay: retraso de la compañia (minutos)
* WeatherDelay: retraso por condiciones atmosféricas (minutos) 
* NASDelay: retraso devido al sistema nacional de aviación (National Airspace System) (minutos)
* SecurityDelay: retraso por cuestiones de seguridad (minutos)
* LAteAircraftDelay: retraso por aviones que llegan tarde (minutos)

**Cancelaciones**
* Cancelled: vuelo cancelado, 1=sí, 0=no
* CancellationCode: código de cancelación

Como hay que dividir distancia / tiempo no pueden haber valores que sean 0 o Nan. Eliminamos esas columnas.

```python
df_reducido = df_reducido.dropna(subset=['AirTime'])
```

```python
df_reducido.drop( df_reducido[df_reducido['AirTime']== 0].index, inplace = True)
round(df_reducido.describe())
```

```python
df_reducido.loc[:, "VelMed"] = round(df_reducido['Distance']/ (df_reducido['AirTime'] / 60))
df_reducido
```

```python
df_reducido.loc[:, "Retard"]= df_reducido['ArrDelay'].apply(lambda x: 'No' if x < 0 else 'Sí')
df_reducido
```
