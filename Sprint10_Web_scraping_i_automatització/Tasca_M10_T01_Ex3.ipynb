```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import scrapy
```

# Selenium

```python
# Configuración de Selenium para ejecución en modo headless e incógnito
options = Options()
options.add_argument('--headless')
options.add_argument('--incognito')

# Inicialización del controlador de Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Obtener la página web
driver.get('https://es.wikipedia.org/wiki/Hapkido#Grados_KUPS_seg%C3%BAn_el_Universal_Hapki-do_DEAMYDC')

# Obtener el contenido de la página después de esperar a que se cargue completamente
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')

# Encontrar la tabla por su clase CSS
table = soup.find('table', {'class': 'wikitable'})

# Extraer datos de la tabla
data = []
if table:
    rows = table.find_all('tr')
    for row in rows:
        cols = row.find_all(['td', 'th'])
        cols = [col.text.strip() for col in cols]
        data.append(cols)

# Crear un DataFrame con los datos extraídos
df = pd.DataFrame(data[1:], columns=data[0])

# Cerrar el navegador
driver.quit()

df
```
