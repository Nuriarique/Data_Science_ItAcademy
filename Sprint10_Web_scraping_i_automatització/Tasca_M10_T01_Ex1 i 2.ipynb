#  BeautifulSoup - quotes

```python
URL = 'http://quotes.toscrape.com/tag/humor/'
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
soup
```

```python
quotes = soup.find_all("div", class_="quote")

for quote in quotes:
    print(quote.prettify(), end="\n"*2)
```

ahora accedemos unicamente al texto, sin etiquetas HTML

```python
for quote in quotes:
    cita = quote.find('span', class_ ='text').text
    autor=quote.find("small", class_="author").text
    print('Cita:', cita)
    print('Autor:', autor)
    print()
```

```python
if page.status_code == 200:
   
    soup = BeautifulSoup(page.text, 'html.parser')

    autores = soup.find_all('small', class_='author')
    citas = soup.find_all('span', class_='text')
    
        
    for autor,cita in zip(autores, citas):
        cita = quote.find('span', class_ ='text').text
        autor=quote.find("small", class_="author").text
        print('Cita:', cita)
        print('Autor:', autor)
        print()

else:
    print(f'Couldn\'t retrieve the page. Status code: {page.status_code}')
```

```python
df = pd.DataFrame({'Autor': [autor.text for autor in autores],
                   'Cita': [cita.text for cita in citas]})

df
```

# Selenium

```python
options = Options()
options.add_argument('--headless')
options.add_argument('--incognito')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get('http://quotes.toscrape.com/tag/humor/')


elems = driver.find_elements(By.CLASS_NAME, 'quote')

autores=[]
citas=[]

for elem in elems:
    autor = elem.find_element(By.CLASS_NAME, "author").text
    autores.append(autor)
    
    cita= elem.find_element(By.CLASS_NAME, "text").text
    citas.append(cita)
    
    print('Cita:', cita)
    print('Autor:', autor)
    print()

driver.quit()
```

```python
#crear df

df_sel = pd.DataFrame({
    'Autor': autores,
    'Cita': citas})

df_sel
```

# Selenium - bolsa

```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

from webdriver_manager.chrome import ChromeDriverManager

options = Options()
options.add_argument('--headless')
options.add_argument('--incognito')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get('https://www.bolsasymercados.es/bme-exchange/es/Negociar/Participantes/Miembros-Renta-Variable')

table =driver.find_element(By.TAG_NAME, 'table')
#print(table)

rows = driver.find_elements(By.CSS_SELECTOR, 'tr[role="row"]')

data = []
for row in rows:
    row_data = []
    cells = row.find_elements(By.TAG_NAME, 'td')
    for cell in cells:
        row_data.append(cell.text.strip())
    data.append(row_data)

    
# Columnas
headers = ['Nombre', 'Código', 'Dirección Social', 'Bolsa']

# df con datos
df = pd.DataFrame(data, columns=headers)

# Eliminar la primera fila 
df = df.iloc[1:]

driver.quit()

df
```
