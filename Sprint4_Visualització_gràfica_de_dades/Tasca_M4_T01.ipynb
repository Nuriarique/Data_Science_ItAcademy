En la columna 'Preu' hay 44 registros con '--'. 

al visualizar solo los registros con '--' se ve que los datos faltantes coinciden en 'trimestre , districto y barrio' tanto en el tipo de 'Lloguer mitjà mensual'  y el 'Lloguer mitjà per superfície'

Deducimos que son datos que no se han generado como no podemos recuperarlos habrá que borrar esos registros para evitar posibles errores. 

```python
df.drop(df[(df['Preu'] == '--')].index, inplace=True)
df
```

```python
df['Preu'] = df['Preu'].astype(float)
df.info()
```

##### separar datos de la columna 'Lloguer_mitja

```python
df_mensual = df.copy()
df_mensual = df_mensual.drop(df_mensual[df_mensual['Lloguer_mitja']=='Lloguer mitjà per superfície (Euros/m2 mes)'].index)
df_mensual
```

```python
df_superficie = df.copy()
df_superficie = df_superficie.drop(df_superficie[df_superficie['Lloguer_mitja']=='Lloguer mitjà mensual (Euros/mes)'].index)
df_superficie
```

### Crea almenys una visualització per:

* Una variable categòrica (Districte o Barri)

```python
barris = df_superficie[['Nom_Districte','Nom_Barri']].drop_duplicates(subset = 'Nom_Barri').groupby('Nom_Districte').count()
barris.plot(kind='bar', figsize=(10, 5))
plt.title("Nº barris per districte")
plt.xlabel('Districtes')
plt.savefig('NumeroDeDistritosPorBarrio.jpg')
```

El gráfico nos muestra el número de barrios que hay en cada districto. Vemos que el mínimo de barrios por districto es 2 y el máximo 10.

Los districtos con mas barrios: Horta-Guinardó, Nou Barris y Sant Martí
Districto con menos: Les corts

* Una variable numèrica (Preu)

```python
df_superficie['Preu'].plot(kind='hist')
plt.title("Frecuencia Preu m2")
plt.xlabel('Preu m2')
plt.savefig('FrecuenciaPreuM2.jpg')
```

los precio por m2 más comunes son entre 12€/m2 y 14€/m2
promedio 13€/m2

```python
df_superficie.boxplot(column=['Preu'], by="Nom_Districte",rot=45)
#df_superficie[['Nom_Districte','Preu']].plot(x='Nom_Districte', y='Preu',kind='scatter', figsize=(15, 10))
#plt.title("Preu m2 per Districte")
#plt.xlabel('Districtes')
#plt.ylabel('Preu m2')
#plt.savefig('PreuM2Districte.jpg')
```

En los districtos de Horta, Sant Andreu, Sants y Sarrià podemos ver que hay valores máximos muy altos o mínimos muy bajos por lo es posible que existan valores raros.

 Los precios medios de Ciutat Vella, Example, Gràcia, Horsta, Nou Barris, Sant Martí y Sants son similares (entre 12 y 14) pero los precios de Sant Martí son más variables que los otros Districtos.

* Dues variables numèriques (Any o Trimestre i Preu)

```python
df_superficie[['Preu', 'Trimestre']].groupby('Trimestre').mean().plot( kind='line')
plt.title("Preu m2 per trimestre")
plt.xlabel('Trimestre')
plt.ylabel('Preu m2')
plt.savefig('PreuM2PerTrimestre.jpg')
```

Vemos la evolución del precio en función del trimestre. 

En el segundo trimestre los precios bajan sin embargo el tercer y el cuarto trimestre los precios se elevevan considereblemente

* Tres variables (Barri o Districte, Trimestre i Preu)

```python
datos = df_superficie.pivot_table(
                        index= 'Trimestre',
                        columns='Nom_Districte',
                        values = 'Preu')

datos.plot(kind='bar', figsize=(18, 9),width = 0.9)
plt.grid(axis = 'y')
plt.title("Mitja del preu m2 per trimestre i districte")
plt.xlabel('Trimestre')
plt.ylabel('Preu m2')
plt.legend(ncol=5, loc='upper center')
plt.savefig('PreuM2PerTrimestreDistricte.jpg')
```

Podemos observar que los precios por m2 son bastante estable en todos los districto.  No se observan grandes variaciones en función del trimestre de año.

Lo que también se destaca son dos districtos con los precios por m2 mas bajo (por debajo de 12€/m2):  Nou barris y Sant Andreu 

Además Sarrià-Sant gervasi es el districto que en todos los trimestre esta con los precios más altos de todos los districtos(por encima de 14€/m2)

El resto de barrios estan entre los 10-12€/m2
